Quarteto de Anscombe é o nome dado a quatro conjuntos de dados que aparentam ser idênticos quando descritos por certas técnicas de estatística descritiva (como a média e a variância), mas que são muito distintos quando exibidos graficamente. Ele leva o nome do estatístico F.J. Anscombe que o publicou pela primeira vez em 1973, com o objetivo de demonstrar tanto a importância de se visualizar os dados antes de analisá-los quanto o efeito dos outliers nas propriedades estatísticas.

Podemos verificar esses datasets no R:

```r
data(anscombe) # baixando o dataset
```

```r
anscombe # visualisando o dataset
```

```r
# verificando o grau de correlação linear
# entre as colunas x1 e y1
cor(anscombe$x1, anscombe$y2) 
```

```r
# Podemos dizer que tem alta correlação linear
# vamos ver a cara dessa distribuição

# configurando tamanho da imagem
options(repr.plot.width=4, repr.plot.height=4)

# gráfico de dispersão 
plot(anscombe$x1, anscombe$y1)
```

Podemos notar uma certa linearidade na dispersão das variáveis acima como esperado.

```r
# Vamos verificar a correlação entre as colunas x2 e y2
cor(anscombe$x2, anscombe$y2)
```

```r
# Obtivemos praticamente o mesmo grau de correlação 
# Será que obteremos um gráfico de dispersão similar
# ao anterior? Vejamos.
options(repr.plot.width=4, repr.plot.height=4)
plot(anscombe$x2, anscombe$y2)
```

Vemos que não há uma correlação linear. 

```r
# Correlação entre x3 e y3
cor(anscombe$x3, anscombe$y3)
```

```r
options(repr.plot.width = 4, repr.plot.height = 4)
plot(anscombe$x3, anscombe$y3)
```

```r
# correlação x4 e y4
cor(anscombe$x4, anscombe$y4)
```

```r
options(repr.plot.width=4, repr.plot.hieght=4)
plot(anscombe$x4, anscombe$y4)
```

Há ainda outros datasets com correlação linear alta mas com gráficos de dispersão bizarros. O mais bizarro deles é o Datasaurus, isso mesmo que você está pensando, ele tem um formato de dinosauro. Para isso, vamos baixar esse dataset no GitHub:

Há ainda outros datasets com correlação linear alta mas com gráficos de dispersão bizarros. As formas incluem uma estrela, e uma cruz, e o "DataSaurus" (criado primeiramente por Alberto Cairo). Os autores publicaram um conjunto de dados que chamam de "DataSaurus Dozen" (também disponível como um pacote R no GitHub, com agradecimentos a Steph Locke) dos 12 diagramas de dispersão mostrados. Curiosamente, os 12 datasets mantêm as mesmas estatísticas de resumo com duas casas decimais. Para ver esses gráficos, vamos baixar esse dataset no GitHub:

```r
devtools::install_github("stephlocke/datasauRus")
```

```r
# Iremos usar o ggplot
# Para plotar esses 12 gráficos datasets
library(ggplot2)
library(datasauRus)
```

```r
ggplot(datasaurus_dozen, aes(x=x, y=y, colour=dataset))+
  geom_point()+
  theme_void()+
  theme(legend.position = "none")+
  facet_wrap(~dataset, ncol=3)
```

Moral da história? Não confie apenas nas estatísticas descritivas, sempre dê uma olhada no gráfico.
